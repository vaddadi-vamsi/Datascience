```python
def open_dataset(file_name):
    opened_file = open(file_name,encoding='utf-8')
    from csv import reader
    read_file = reader(opened_file)
    data = list(read_file)
    
    return data

def explore_data(dataset, start, end, rows_and_columns=False):
    dataset_slice = dataset[start:end]    
    for row in dataset_slice:
        print(row)
        print('\n') # adds a new (empty) line after each row

    if rows_and_columns:
        print('Number of rows:', len(dataset))
        print('Number of columns:', len(dataset[0]))
        
apple_dataset=open_dataset('C:/Users/vadda/PycharmProjects/Datascience/AppleStore.csv')
google_dataset=open_dataset('C:/Users/vadda/PycharmProjects/Datascience/googleplaystore.csv')
apple_dataset_without_header=apple_dataset[1:]
google_dataset_without_header=google_dataset[1:]
i=0
for app in google_dataset_without_header:
    i=i+1
    if (app[3] == "3.0M"):
        print(i)

apple_data=explore_data(apple_dataset[1:],0,2,True)
google_data=explore_data(google_dataset[1:],0,2,True)

print(apple_dataset[:3])

print('\n',google_dataset[0],google_dataset[10473])
del google_dataset[10473]
print('\n',google_dataset[0],google_dataset[10473])
del google_dataset_without_header[10472]
print('\n',google_dataset[0],google_dataset_without_header[10472])
```

```python
def duplicate_app(dataset,list=False):
    duplicate_app=[]
    unique_app=[]
    for app in dataset:
        name=app[0]
        if name in unique_app:
            duplicate_app.append(name)
        else:
            unique_app.append(name)
    print("Total number of Unique apps", len(unique_app))
    print("Total number of Duplicate apps", len(duplicate_app))
    print("Examples of duplicate apps",explore_data(duplicate_app,1,len(unique_app)))
    print('\n')

duplicate_app(apple_dataset)
duplicate_app(google_dataset)
    
```

```python
def duplicates_count(dataset):
    reviews_max={}
    for app in dataset:
        name = app[0]
        n_reviews= float(app[3])
        if name in reviews_max and reviews_max[name] < n_reviews:
            reviews_max[name]= n_reviews
        elif name not in reviews_max:
            reviews_max[name]=n_reviews
    print("Expected length after duplicates", len(reviews_max))
    return reviews_max
reviews_final=duplicates_count(google_dataset[1:])
```

```python
print (type(reviews_final))
android_clean = []
already_added = []
for app in google_dataset_without_header:
    name =  app[0]
    n_reviews= float(app[3])
    if n_reviews == reviews_final[name] and name not in already_added:
        android_clean.append(app)
        already_added.append(name)
print(already_added)
      
```

```python
def is_english(string):
    for character in string:
        if ord(character) > 127:
            return False
        else:
            return True
print(is_english('Instagram'))
print(is_english('爱奇艺PPS -《欢乐颂2》电视剧热播'))
print(is_english('Docs To Go™ Free Office Suite'))
print(is_english('Instachat 😜'))
        
        
```

```python
def is_english1(string):   
    count=0
    for character in string:
        if ord(character) > 127:
            count+=1
    if count > 3:
        return False
    else:
        return True
print(is_english1('Instagram'))
print(is_english1('爱奇艺PPS -《欢乐颂2》电视剧热播'))
print(is_english1('Docs To Go™ Free Office Suite'))
print(is_english1('Instachat 😜'))
```

```python
google_english = []
apple_english = []
for app in apple_dataset_without_header:
    name = app[1]
    if is_english1(name):
        apple_english.append(app)
for app in android_clean:
    name = app[0]
    if is_english1(name):
        google_english.append(app)
print('\n',apple_dataset[0],'\n')
explore_data(apple_english,0,2,True)
print("\n")
print('\n',google_dataset[0],'\n')
explore_data(google_english,0,2,True)
        
```

```python
def is_free(dataset,position):
    free = []
    for app in dataset:
        name=app[position]
        if name == '0' or name == '0.0':
            free.append(app)
    return free
apple_final=is_free(apple_english,4)
google_final=is_free(google_english,7)

print('\n',apple_dataset[0],'\n')
explore_data(apple_final,0,2,True)
print("\n")
print('\n',google_dataset[0],'\n')
explore_data(google_final,0,2,True)
```
